```python
#Datewise number of picking and placing activity done.
total_num_acivity = data.groupby('date')['activity'].value_counts().reset_index()
```

```python
data = data.sort_values(['date','time'])
```

```python
data
```

```python
#datatime column
data['datetime'] = pd.to_datetime(data['date'].astype(str) + ' ' + data['time'].apply(lambda t: t.strftime('%H:%M:%S')))
```

```python
data.head()
```

```python
data['duration'] = data['datetime'].diff().shift(-1)
```

```python
data
```

```python
data
```

```python
data['duration'] = data['duration'].fillna(pd.Timedelta(seconds=0))
```

```python
#data.to_excel('Each_position_duration.xlsx', index=False)
```

```python
inside_outside_duration = data.groupby(['date', 'position'])['duration'].sum().reset_index()
```

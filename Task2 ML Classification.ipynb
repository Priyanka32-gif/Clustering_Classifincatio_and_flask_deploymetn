```python
# Function to predict the lables and explain it
def predict_and_explain(data_point, rf_classifier, scaler, label_encoder):
    
    data_point_scaled = scaler.transform([data_point])
    prediction = rf_classifier.predict(data_point_scaled)[0]
    predicted_label = label_encoder.inverse_transform([prediction])[0]
    
    explanation = f"The data point is predicted to belong to class {predicted_label} because:\n"
    feature_importances = rf_classifier.feature_importances_
    
    for feature, importance in zip(X.columns, feature_importances):
        explanation += f"Feature {feature} importance: {importance:.4f}\n"
    
    return predicted_label, explanation
```

```python
#tesing the function
new_data_point = [0.5, 0.7, 0.3, 0.8, 0.4, 0.9, 0.2, 0.5, 0.9, 0.1, 0.6, 0.5, 0.8, 0.6, 0.3, 0.1, 0.4, 0.7]

# Predict and explain
predicted_label, explanation = predict_and_explain(new_data_point, rf_classifier, scaler, le)

print(f"Predicted Class: {predicted_label}")
print("Explanation:")
print(explanation)
```
